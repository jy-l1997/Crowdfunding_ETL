```python
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
```

# Extract crowdfunding.xlsx data

```python
# Read data into Pandas Dataframe
crowdfunding_info_df = pd.read_excel('crowdfunding.xlsx')
crowdfunding_info_df.head()
```

```python
# Create a copy of the crowdfunding_info_df DataFrame and name it campaign_df

campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Rename the blurb, launched_at, and deadline columns

campaign_df.rename(columns={'blurb': 'description', 'launched_at': 'launch_date', 'deadline': 'end_date'}, inplace=True)
```

```python
campaign_df.head()
```

```python
# Change the goal column to a float data type

campaign_df['goal'] = campaign_df['goal'].astype(float)
```

```python
campaign_df.head()
```

```python
# Convert the pledged column to a float data type 
campaign_df['pledged'] = campaign_df['pledged'].astype(float)
```

```python
# check datatypes

data_types = campaign_df.dtypes
print(data_types)

```

```python
# Format the launched_date and end_date columns to datetime format
from datetime import datetime as dt
campaign_df['launch_date'] = pd.to_datetime(campaign_df['launch_date'], utc=True)
campaign_df['end_date'] = pd.to_datetime(campaign_df['end_date'], utc=True)
```
